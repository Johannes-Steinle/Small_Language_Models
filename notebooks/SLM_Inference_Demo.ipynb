# SLM Inferenz-Demo

Dieses Notebook konzentriert sich auf das **Ausführen** des Modells (Inferenz).
Nützlich zum Testen von Prompts oder eines fine-getunten Adapters.

```python
import torch
from transformers import pipeline

# 4B Instruct-Modell für Chat-Fähigkeiten
model_id = "google/gemma-3-4b-it"

pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.bfloat16,
        "quantization_config": {"load_in_4bit": True}
    },
    device_map="auto",
)

messages = [
    {"role": "user", "content": "Erkläre Quantenphysik für ein 5-jähriges Kind."},
]

outputs = pipe(messages, max_new_tokens=256)
print(outputs[0]["generated_text"][-1]["content"])
```
