# SLM Inference Demo

This notebook focuses purely on **running** the model (Inference).
Useful to test prompts or a fine-tuned adapter.

```python
import torch
from transformers import pipeline

# Using the 4B Instruct model for chat capabilities
model_id = "google/gemma-3-4b-it"

pipe = pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={
        "torch_dtype": torch.bfloat16,
        "quantization_config": {"load_in_4bit": True}
    },
    device_map="auto",
)

messages = [
    {"role": "user", "content": "Explain Quantum Physics to a 5 year old."},  
]

outputs = pipe(messages, max_new_tokens=256)
print(outputs[0]["generated_text"][-1]["content"])
```
